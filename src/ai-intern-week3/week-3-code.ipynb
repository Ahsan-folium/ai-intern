````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
````

````python
wine_df = pd.read_csv('winequality-red.csv' , sep=';')
````

## EDA

````python
wine_df
````

````python
wine_df.info()
````

````python
wine_df.describe()
````

````python
wine_df['good'] = (wine_df['quality'] >= 6).astype(int)   
# made into 2 cattegories for better training
````

````python
wine_df
````

````python
X = wine_df.drop(['quality', 'good'], axis=1)    # features
y = wine_df['good']     # label
````

````python
y
````

````python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
````

## Logistic Regression

````python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)
y_predict_lr = lr.predict(X_test)
````

## Decision Tree 

````python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
````

## Metrics

````python
# precison , recall , F1
from sklearn.metrics import classification_report

print("Logistic Regression:\n", classification_report(y_test, y_predict_lr))
````

````python
print("Decision Tree:\n", classification_report(y_test, y_pred_dt))
````

```Decision Tree gives very slightly better accuracy```

````python
from sklearn.model_selection import cross_val_score

scores_lr = cross_val_score(lr, X_scaled, y, cv=5, scoring='f1')

print("LogReg F1 scores:", scores_lr, "Mean:", scores_lr.mean(), "Std:", scores_lr.std())
````

````python

scores_dt = cross_val_score(dt, X_scaled, y, cv=5, scoring='f1')
print("Decision Tree F1 scores:", scores_dt, "Mean:", scores_dt.mean(), "Std:", scores_dt.std())
````
